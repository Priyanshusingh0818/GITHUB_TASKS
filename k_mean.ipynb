```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import silhouette_score
```

```python
np.random.seed(42)
n_student=200
grades=np.random.randint(40,100,size=(n_student,3))
df=pd.DataFrame(grades,columns=['phy','chem','math'])
```

```python
df.head()
```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
clusters=kmeans.fit_predict(df)
df['Cluster']=clusters
df.head()
```

```python
inertia = kmeans.inertia_
silhouette_avg = silhouette_score(df, clusters)
print(f'Inertia: {inertia}')
print(f'Silhouette Score: {silhouette_avg:.2f}')
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(df['math'],df['phy'],df['chem'],c=df['Cluster'],cmap='viridis',s=50)
ax.set_xlabel('math grades')
ax.set_ylabel('phy grades')
ax.set_zlabel('chem grades')
ax.set_title('KMeans cluster of student by gradews')
plt.show()
```
